```python
import json
import csv

# 入出力ファイル
input_file = "../data/koji_after_rag_8.jsonl" 
output_file = "../data/csv/koji_after_rag.csv" 

# JSONLを一度すべて読み込み（retrrieved数を確認するため）
with open(input_file, 'r', encoding='utf-8') as f_jsonl:
    data_list = [json.loads(line) for line in f_jsonl]

# 最大retrieved数を確認
max_retrieved = max(len(item.get("retrieved", [])) for item in data_list)

# ヘッダー作成
header = ["instruction", "input", "output"]
for i in range(1, max_retrieved + 1):
    header.extend([
        f"ret_{i}_text",
        f"ret_{i}_source_title",
        f"ret_{i}_source_url",
        f"ret_{i}_chunk_id",
        f"ret_{i}_position",
        f"ret_{i}_is_official"
    ])

# CSV書き込み
with open(output_file, 'w', encoding='utf-8', newline='') as f_csv:
    writer = csv.writer(f_csv)
    writer.writerow(header)

    for item in data_list:
        row = [
            item.get("instruction", ""),
            item.get("input", ""),
            item.get("output", "")
        ]

        retrieved_list = item.get("retrieved", [])
        for r in retrieved_list:
            text = r.get("text", "")
            meta = r.get("meta", {})
            row.extend([
                text,
                meta.get("source_title", ""),
                meta.get("source_url", ""),
                meta.get("chunk_id", ""),
                meta.get("position", ""),
                meta.get("is_official", "")
            ])
        
        #  不足分を空欄で埋める（retrievedが少ない行用）
        num_missing = (max_retrieved - len(retrieved_list)) * 6
        row.extend([""] * num_missing)

        writer.writerow(row)

print(f"✅ CSV出力完了: {output_file}")
print(f"最大retrieved数: {max_retrieved}")
```
